# Connecting (Tunneling) Jupyter notebook from a Cluster (this case Anthill-SDSU) to your MAC

You will need access to a cluster to do this. I will use the SDSU Anthill cluster, but this will work with any cluster. BUT, you must have all the dependencies for the jupyter notebook inside your cluster account (e.g., anaconda, python, R , jupyter notebook)

1. Log in to your antill account:
    USER@anthill.sdsu.edu
2. If  all your dependencies are already in the anthill account then- go ahead and open a jupyter notebook within the cluster:
    [torres@anthill ~]$ jupyter notebook --no-browser --port 4444 &
    
wait a bit and something like this should pop up:
    [1] 6813
[torres@anthill ~]$ [I 12:24:12.220 NotebookApp] Serving notebooks from local directory: /usr/lustrefs/anthill/homes/home3/torres
[I 12:24:12.220 NotebookApp] 0 active kernels 
[I 12:24:12.220 NotebookApp] The IPython Notebook is running at: http://localhost:4444/
[I 12:24:12.220 NotebookApp] Use Control-C to stop this server and shut down all kernels (twice to skip confirmation).

now you have a local host port number, here it is 4444 yours will be different, depending on the port number you decided to use.

3. Open a new tab on YOUR computers terminal and now to connect to the clusters jupyter notebook like this:

    ssh -NL 4444:localhost:4444 torres@anthill.sdsu.edu
   
should then ask for your anthill passsword, then your terminal will look like this:
    
    Pedros-MacBook-Pro-3:BioinformaticsBC Pedro_Torres$ ssh -NL 4444:localhost:4444 torres@anthill.sdsu.edu
torres@anthill.sdsu.edu's password: 

nothing else will happen the screen will stay as above, if no errors pop up then you are good.

4. Go ahead and copy the localhost directory where the ipython notebook is running within the cluster and paste it on your browser. 
    http://localhost:4444/
 
5. Jupyter notebook is good to go.

    
    

# Tunneling to Windows

##### You will need to install PuTTY onto you computer http://www.putty.org/

###### 1. Create a new putty session
  So that you only need to have one Putty session open, we'll make a new TSCC Session. Create one with username@anthill.sdsu.edu, and call the session "Anthill Jupyter"
###### 2. Add your private key and allow forwarding
 Go to "Connection > SSH > Auth." Click the checkbox next to "Allow agent forwarding" and add the "Putty Private Key" that you created with the biom262_rsa file.
###### 3. Add a tunnel
Go to "Connnection > SSH > Tunnels." Then:

Click the checkbox next to "Local ports accept connections from other ports"
Add your #### for your source port
Add localhost:#### for your Destination
Click "Local"
Click "Add"
###### 4. Save settings
So you don't have to do this every time... Save your settings! Go all the way back to the "Session" window and click "Save"
