```python
# Pedro J. Torres
import pandas as pd
import sqlite3
# Output is from https://github.com/medema-group/bigslice input folder example authors supply
# For more information regarding the db schema look at the authors supplementary file ; specifically page 42
# foundn intheri paper https://academic.oup.com/gigascience/article/10/1/giaa154/6092777

con = sqlite3.connect("data.db")
```

```python
#Basic info on BGCs ran thorugh bigslice using example input folder they supply
bgc_info = f"""\
Select dataset.name, bgc.orig_folder, bgc.orig_filename, bgc.on_contig_edge, bgc.length_nt 
from bgc,dataset 
where bgc.dataset_id=dataset.id order by bgc.id;
"""

surveys_df = pd.read_sql_query(bgc_info, con)
surveys_df
```

```python
# BGC taxonomy
    
taxinfo = f"""\
select bgc_taxonomy.bgc_id, taxon.name from bgc_taxonomy, taxon 
where taxon.id=bgc_taxonomy.taxon_id and taxon.level=7;
"""

tax_df = pd.read_sql_query(taxinfo, con)
tax_df
```

```python
# Get BGC, checm class and subclass as well as cds products and which regions they are found in 
info = f"""\
Select * from bgc
INNER JOIN bgc_class 
ON bgc_class.bgc_id=bgc.id
INNER JOIN chem_subclass as  cs
ON cs.id=bgc_class.chem_subclass_id
INNER JOIN chem_class 
ON chem_class.id=cs.class_id
INNER JOIN cds
ON cds.bgc_id=bgc.id;
"""

surveys_df = pd.read_sql_query(info, con)
surveys_df
```

```python
hge_info = f"""\
Select * from bgc
INNER JOIN bgc_class 
ON bgc_class.bgc_id=bgc.id
INNER JOIN chem_subclass as  cs
ON cs.id=bgc_class.chem_subclass_id
INNER JOIN chem_class 
ON chem_class.id=cs.class_id
INNER JOIN taxon 
ON taxon.id=bgc.id
INNER JOIN taxon_class 
ON taxon_class.level=taxon.level;
"""

surveys_df = pd.read_sql_query(hge_info, con)
surveys_df
```

```python
hge_info = f"""\
Select bgc.name, cs.name,chem_class.name, cds.* from bgc
INNER JOIN bgc_class 
ON bgc_class.bgc_id=bgc.id
INNER JOIN chem_subclass as  cs
ON cs.id=bgc_class.chem_subclass_id
INNER JOIN chem_class 
ON chem_class.id=cs.class_id
INNER JOIN cds
ON cds.bgc_id=bgc.id
;
"""

surveys_df = pd.read_sql_query(hge_info, con)
surveys_df
```

```python
cheminfo = f"""\
select chem_class.name AS class, chem_subclass.name as subclass from chem_subclass, chem_class
where chem_class.id=chem_subclass.class_id;
"""

chem_df = pd.read_sql_query(cheminfo, con)
chem_df


```
